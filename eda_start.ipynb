```python
payments = pd.read_csv(r"C:\\Users\\acer\\OneDrive\Documents\\Ecommers_ Analysis_Project\\data\\raw\\olist_order_payments_dataset.csv")
payments.head()
```

```python
payments.info()
```

```python
full_data = pd.merge(cust_orders, payments, on = 'order_id' , how = 'inner')
full_data.shape
```

```python
full_data.groupby('customer_state')['payment_value'].sum().sort_values(ascending= False).head(10)
```

```python
import matplotlib.pyplot as plt
full_data.groupby('customer_state')['payment_value'].sum().sort_values(ascending= False).head(10).plot(kind = 'bar')
plt.title("Top 10 State by Revenue")
plt.show()
```

```python
full_data.groupby('customer_city') ['payment_value'].sum().sort_values(ascending = False).head(10).plot(kind ='bar')
plt.title("Top 10  City by Revenue")
plt.show()
```

```python
full_data['order_purchase_timestamp']= pd.to_datetime(full_data['order_purchase_timestamp'])
full_data['month'] = full_data['order_purchase_timestamp'].dt.to_period('M')
full_data.groupby('month')['payment_value'].sum().plot()
plt.title("Monthly Revenue Trend")
plt.show()
```
